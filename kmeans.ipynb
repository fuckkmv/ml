```python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
```

```python
iris = load_iris()
```

```python
x = iris.data
y = iris.target
```

```python
y
```

```python
plt.scatter(x.T[2], x.T[3])
```

```python
df = pd.DataFrame(x)
```

```python
df['result'] = y
```

```python
df1 = df[df['result']==0]
df2 = df[df['result']==1]
df3 = df[df['result']==2]
```

```python
plt.scatter(df1[2],df1[3])
plt.scatter(df2[2],df2[3])
plt.scatter(df3[2],df3[3])
```

```python
km = KMeans(n_clusters = 3, n_init='auto', init = 'k-means++')
```

```python
y_predicted = km.fit_predict(df[[2,3]])
y_predicted
```

```python
km.cluster_centers_
```

```python
df['k-means-pred'] = y_predicted
```

```python
df1 = df[df['k-means-pred']==0]
df2 = df[df['k-means-pred']==1]
df3 = df[df['k-means-pred']==2]
plt.scatter(df1[2],df1[3])
plt.scatter(df2[2],df2[3])
plt.scatter(df3[2],df3[3])
```

```python
y_predicted = np.select([y==1,y==0], [0,1], y_predicted)
```

```python
print(confusion_matrix(y,y_predicted))
```
