```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
import numpy as np
import pandas as pd
```

```python
iris = load_iris()
x = iris.data
y = iris.target
```

```python
deci_tree = DecisionTreeClassifier(criterion = 'entropy')
```

```python
kf = KFold(n_splits = 10, shuffle = True)
```

```python
score = cross_val_score(deci_tree,x,y,cv = kf)
```

```python
score = np.array(score)
```

```python
score
```

```python
score.mean()*100
```

```python
# below - for generating decision tree, but using holdout
# only above is for cross validation
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.25)
```

```python
deci_tree.fit(x_train, y_train)
```

```python
tree.plot_tree(deci_tree,
                class_names = ['setosa','versicolor','virginica'],
                filled = True)
```

```python
predictions = deci_tree.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print(confusion_matrix(y_test, predictions))
```

```python
print(classification_report(y_test, predictions))
```
